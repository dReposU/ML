**Algoritmo A priori**   
Utilizaremos la libreria mi¿lxtend.
Identificar reglas de asociacion en transaccciones productos informatico

```python
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings("ignore", category=DeprecationWarning) #Ignora Warnings
```

```python
def proceso_apriori():

  # 1 Creacion de Dataset
  dataset = [
      ['Laptop', 'Mouse', 'Monitor'],
      ['Laptop'], ['Mouse'],
      ['Mouse', 'Teclado'],
      ['Laptop', 'Monitor', 'Teclado'],
      ['Laptop', 'Mouse'],
      ['Mouse', 'Teclado'],
      ['Laptop', 'Teclado'],
      ['Laptop', 'Mouse', 'Monitor', 'Teclado'],
      ['Teclado', 'Monitor']
  ]

  # 2 Preprocesamiento: Convertir a formato One-Hot Encoding
  from mlxtend.preprocessing import TransactionEncoder

  te = TransactionEncoder()
  te_ary = te.fit(dataset).transform(dataset)
  df = pd.DataFrame(te_ary, columns=te.columns_)

  # 3. Aplicar el algoritmo Apriori
  frequent_itemsets = apriori(df, min_support=0.3, use_colnames=True)

  #4. Generar las Reglas de Asociacion
  rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.7)

  print("--- Conjunto de Datos Frecuentes---")
  print(frequent_itemsets)

  print("\n--- Reglas de Asociacion---")
  #print(rules)
  print(rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']])
```

```python
proceso_apriori()
```
