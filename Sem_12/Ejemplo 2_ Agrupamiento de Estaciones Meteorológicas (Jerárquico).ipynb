**Objetivo**: Entender cómo se relacionan zonas climáticas según temperatura y humedad.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage

def agrupamiento_climatico():
    """
    Uso de agrupamiento jerarquico para visualizar relaciones climaticas.
    """
    # [Temp_Media, Humedad_Relativa]
    zonas = np.array([[30, 80], [32, 85], [10, 20], [12, 25], [22, 50], [25, 55]])
    ciudades = ['Selva A', 'Selva B', 'Desierto A', 'Desierto B', 'Valle A', 'Valle B']

    # Generar el enlace (linkage)
    Z = linkage(zonas, 'ward')

    # Dibujar Dendrograma
    plt.figure(figsize=(8, 4))
    dendrogram(Z, labels=ciudades)
    plt.title('Dendrograma de Similitud Climática')
    plt.ylabel('Distancia Euclídea')
    plt.show()

agrupamiento_climatico()
```
