```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
horas_estudio = np.array([2, 3, 4, 5, 6]).reshape(-1,1) # X
calificaciones = np.array([70, 75, 80, 85, 90])         # Y
```

```python
modelo = LinearRegression()
modelo.fit(horas_estudio, calificaciones)
```

```python
Calificaciones_pred = modelo.predict(horas_estudio)
```

```python
df = pd.DataFrame({
    "Horas de Estudio": horas_estudio.flatten(),
    "Calificaciones": calificaciones,
    "Calificaciones Pred": Calificaciones_pred
})


plt.figure(figsize=(10,6))
plt.scatter(df["Horas de Estudio"], df["Calificaciones"], color="blue", label="Datos Reales")
plt.plot(df["Horas de Estudio"], df["Calificaciones Pred"], color="red", label="Linea de Regresion")
plt.xlabel("Horas de Estudio")
plt.ylabel("Calificaciones")
plt.title("Regresion Lineal")
plt.legend()
plt.show()
plt.grid()

plt.show()
```
