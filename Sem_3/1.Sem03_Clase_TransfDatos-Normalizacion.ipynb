```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import scipy.stats as ss

from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
```

```python
import sys

if 'google.colab' in sys.modules:
    data = pd.read_csv(f'https://raw.githubusercontent.com/dReposU/ML/main/Sem_2/bupa.csv', sep=";")
else:
    data = pd.read_csv("../Sem_2/bupa.csv", sep=";")
data.head()
```

```python
data.dtypes
```

```python
data["V7"] = data.V7.astype(object)
```

```python
data.dtypes
```

```python
data_X = data.drop("V7", axis=1)
```

```python
data_X.head()
```

**Normalizacion StandardScaler z-core**

```python
data_X_1 = pd.DataFrame(StandardScaler().fit_transform(data_X))
data_X_1
```

```python
data_X_1.columns = ["V1", "V2","V3","V4","V5","V6"]
data_X_1.head()
```

```python
plt.boxplot(data_X_1.loc[:, ["V1", "V2","V3","V4","V5","V6"]])
plt.show()
```

**Normalizacion Min-Max**

```python
data_X_2 = pd.DataFrame(MinMaxScaler().fit_transform(data_X))
data_X_2.columns = ["V1", "V2","V3","V4","V5","V6"]
data_X_2.head()
```

```python
plt.boxplot(data_X_2)
plt.show()
```
