## 1. Importar Librerias 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## 2. Carga de datos

```python
import sys

criptomoneda = 'ETHO'

if 'google.colab' in sys.modules:
    df = pd.read_csv(f'https://raw.githubusercontent.com/dReposU/ML/main/Sem_1/Ejercicio-Preprocesamiento/data/{criptomoneda}-USD.csv')
else:
    df = pd.read_csv(f'data/{criptomoneda}-USD.csv')
df.head()
```

## 3. Exploración Inicial

```python
df.info()
```

```python
df.describe()
```

## 4. Limpieza y Transformación

```python
df["Date"] = pd.to_datetime(df["Date"], format='%Y-%m-%d') # convertir a datetime
```

```python
df.sort_values(by="Date", inplace=True) # Ordenar por fecha y que se aplique al dataframe
```

```python
df.isnull().sum() # verificar nulos para cada campo
```

## 5. Analisis Univariado

Precio de Cierre

```python
plt.figure(figsize=(8,5))
sns.histplot(df["Close"], kde=True)
plt.title("Distribución Del Precio De Cierre")
plt.show()
```

```python
plt.figure(figsize=(8,10))
sns.boxplot(x=df["Close"])
plt.title("Distribución Del Precio De Cierre")
plt.show()
```

```python
plt.figure(figsize=(20,10))
plt.plot(df['Date'], df['Close'], color='red', linestyle='-')
plt.xticks(rotation=45) 
plt.grid(True)
plt.show()
```
